# Análise da variação de ações da Bovespa

```python
import pandas as pd
import plotly.express as px
```

```python
df_principal = pd.read_excel('/content/Imersão Python - Tabela de ações .xlsx', sheet_name = 'Principal')
df_principal.head(10)
```

```python
df_total_acoes = pd.read_excel('/content/Imersão Python - Tabela de ações .xlsx', sheet_name = 'Total_de_acoes')
df_total_acoes.head(10)
```

```python
df_ticker = pd.read_excel('/content/Imersão Python - Tabela de ações .xlsx', sheet_name = 'Ticker')
df_ticker.head(10)
```

```python
df_chatgpt = pd.read_excel('/content/Imersão Python - Tabela de ações .xlsx', sheet_name = 'Chatgpt')
df_chatgpt.head(10)
```

```python
df_principal = df_principal[['Ativo','Data','Último (R$)','Var. Dia (%)']].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns = {'Último (R$)':'valor_final','Var. Dia (%)':'var_dia_pct'}).copy()
df_principal
```

```python
df_principal['var_pct'] = df_principal['var_dia_pct'] / 100
df_principal['valor_inicial'] = df_principal['valor_final'] /(df_principal['var_pct'] + 1)
df_principal
```

```python
df_principal = df_principal.merge(df_total_acoes, left_on = 'Ativo', right_on = 'Código', how = 'left')
df_principal
```

```python
df_principal = df_principal.drop(columns = ['Código'])
df_principal
```

```python
df_principal['variacao_rs'] = (df_principal['valor_final'] - df_principal['valor_inicial'])* df_principal['Qtde. Teórica']
df_principal
```

```python
pd.options.display.float_format = '{:.2f}'.format
df_principal
```

```python
df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)
df_principal
```

```python
df_principal = df_principal.rename(columns = {'Qtde. Teórica':'qtd_teorica'}).copy()
df_principal
```

```python
df_principal['Resultado'] = df_principal['variacao_rs'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Estável'))
df_principal
```

```python
df_principal = df_principal.merge(df_ticker, left_on = 'Ativo', right_on = 'Ticker', how = 'left')
df_principal = df_principal.drop(columns = ['Ticker'])
df_principal
```

```python
df_chatgpt
```

```python
df_principal = df_principal.merge(df_chatgpt, left_on = 'Nome', right_on = 'Empresa', how = 'left')
df_principal = df_principal.drop(columns = ['Empresa'])
df_principal
```

```python
df_principal['cat_idade'] = df_principal['Idade (Anos)'].apply(lambda x: 'Mais de 100 anos' if x > 100 else ('Menos de 50 anos' if x < 50 else 'Entre 50 e 100 anos'))
df_principal
```

```python
# 1. Converta a coluna 'variação_rs' para números
df_principal['variacao_rs'] = df_principal['variacao_rs'].astype(float)

# 2. Calcule os valores solicitados
maior = df_principal['variacao_rs'].max()
menor = df_principal['variacao_rs'].min()
media = df_principal['variacao_rs'].mean()

# 3. Filtrar o DataFrame para calcular os máximos de quem subiu e desceu
maior_subiu = df_principal[df_principal['Resultado'] == 'Subiu']['variacao_rs'].max()
maior_desceu = df_principal[df_principal['Resultado'] == 'Desceu']['variacao_rs'].max()

# 4. Realize os cálculos necessários para cada categoria
resultado_maior = f"Maior R$ {maior:,.2f}"
resultado_menor = f"Menor R$ {menor:,.2f}"
resultado_media = f"Média R$ {media:,.2f}"
resultado_maior_subiu = f"Maior de quem subiu R$ {maior_subiu:,.2f}"
resultado_maior_desceu = f"Maior de quem desceu R$ {maior_desceu:,.2f}"

# Exibir os resultados
print(resultado_maior)
print(resultado_menor)
print(resultado_media)
print(resultado_maior_subiu)
print(resultado_maior_desceu)
```

```python
df_principal_subiu = df_principal[df_principal['Resultado'] == 'Subiu']
df_principal_subiu
```

```python
df_analise_segmento = df_principal_subiu.groupby('Segmento')['variacao_rs'].sum().reset_index()
df_analise_segmento
```

```python
df_analise_saldo = df_principal.groupby('Resultado')['variacao_rs'].sum().reset_index()
df_analise_saldo
```

```python
# Criação do gráfico de barras para a variação do resultado em reais
fig = px.bar(df_analise_saldo, x = 'Resultado', y = 'variacao_rs', text = 'variacao_rs', title = 'Variação Reais por Resultado')
fig.update_traces(texttemplate='R$ %{text:,.0f}', textposition='outside')
fig.update_layout(yaxis=dict(tickformat=',.0f'))
fig.show()
```

```python
# Criação do gráfico de pizza para a variação das ações por segmento

fig = px.pie(values=df_analise_segmento['variacao_rs'], names=df_analise_segmento['Segmento'], title='Variação das ações por segmento')
fig.show()
```

```python
# Gerando o dataframe com a variação por faixa etária

df_analise_idade_var = df_principal.groupby('cat_idade')['variacao_rs'].sum().map('R${:,.2f}'.format).reset_index()
df_analise_idade_cont = df_principal.groupby('cat_idade')['Ativo'].count().reset_index()
df_analise_idade = pd.merge(df_analise_idade_cont,df_analise_idade_var,on='cat_idade')
df_analise_idade
```

```python
# Criação do gráfico de barras para a variação do resultado de acordocom a faixa  etária

fig = px.bar(df_analise_idade, x = 'cat_idade', y = 'variacao_rs', title = 'Variação em Reais de acordo com a idade da empresa')
fig.update_layout(yaxis=dict(tickformat=',.0f'))
fig.show()
```
